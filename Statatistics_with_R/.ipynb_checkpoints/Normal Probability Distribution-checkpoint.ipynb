# Normal Probability Distribution

![](normal_dist.png)

## Characteristics
    * Symmetrically distributed
    * Long tails (infinity tail doesn't meet X-axis, the distance between tail and X-axis will keep reducing but never touch)
    * Bell Shaped
    * Meam Median and Mode are the same
    * Standard Normal Distribution has mean=0, sd=1 and is the default rnorm

## Some Formula

$X ~ N(\mu{}, \sigma{})$

$z = \frac{x-\mu{}}{\sigma}$

$x = \mu{} + z\sigma{}$

$f(x) = \frac{1}{\sigma{}\sqrt{2\pi{}}}.e^{-\frac{1}{2}z}$      <= Probability density function

[$e = 25.718...$]

## TWO factor define the shape of the curve
    * Mean 
    * Standard Deviation

## The Empirical Rule

How many number x exist between one standard deviation from the mean (𝜇-d𝜎, 𝜇+d𝜎)

* (𝜇-1𝜎, 𝜇+1𝜎) => 68%
* (𝜇-2𝜎, 𝜇+2𝜎) => 95%
* (𝜇-3𝜎, 𝜇+3𝜎) => 99.7%

```r
set.seed(118)

n = floor(runif(1, min=118, max=100000))
mu = runif(1, min= -100, max=1000)
std = runif(1, min= 0, max=100)

n = 2000
mu = 74
std = 8
x = rnorm(n, mean=mu, sd=std)

cat("n    [",n,"]\n")
cat("mean [",mu,"]\n")
cat("std  [",std,"]\n\n")

cat("% of x with 1std from mean : ", sum(x>mu-std & x < mu + std)/length(x)*100, "%\n")
cat("% of x with 2std from mean : ", sum(x>mu-2*std & x < mu + 2*std)/length(x)*100, "%\n")
cat("% of x with 3std from mean : ", sum(x>mu-3*std & x < mu + 3*std)/length(x)*100, "%\n")
```

```r
hist(x, c='blue', density=x )
```
